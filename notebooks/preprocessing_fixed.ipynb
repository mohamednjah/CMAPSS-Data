```python
query1 = 'SELECT * FROM "FD001_engine_data";'
query2 = 'SELECT * FROM "FD001_rul_data";'

# Read the SQL query into a DataFrame
df = pd.read_sql_query(query1, con=engine)
RUL = pd.read_sql_query(query2, con=engine)

# Display basic information about the DataFrame
print("Data imported successfully.")
print("DataFrame shape:", df.shape)
df.head()
```

```python
# Step 1: Display dataset information
df.info()

# Step 2: Check for missing values
print("Missing values:\n", df.isnull().sum())

# Step 3: Display basic statistics
df.describe()
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# ✅ Increase threshold to avoid removing all data
threshold = 5  # Try 4 or 5

# Identify numeric columns
numeric_columns = df.select_dtypes(include=[np.number]).columns

# Compute Z-Scores
z_scores = np.abs((df[numeric_columns] - df[numeric_columns].mean()) / df[numeric_columns].std())

# ✅ Fix: Use .any(axis=1) to keep more data
df_cleaned = df[(z_scores < threshold).any(axis=1)].copy()

print(f"Rows before outlier removal: {df.shape[0]}")
print(f"Rows after outlier removal: {df_cleaned.shape[0]}")
```

```python
from sklearn.preprocessing import StandardScaler

# Step 2: Normalize the numeric columns before detecting outliers
scaler = StandardScaler()

# Selecting only numerical columns
numeric_columns = df.select_dtypes(include=[np.number]).columns

# Apply StandardScaler
df_scaled = pd.DataFrame(scaler.fit_transform(df[numeric_columns]), columns=numeric_columns)

# Compute Z-Scores on scaled data
z_scores = np.abs(df_scaled)

# Remove outliers based on Z-Score threshold (still set to 4)
df_cleaned = df[(z_scores < 4).all(axis=1)]

print("Outliers removed. New dataset shape:", df_cleaned.shape)
```

```python
sensor_columns = [col for col in df_cleaned.columns if "sensor" in col]
```

```python
# ✅ Fix: Ensure all sensor columns are float32 before applying MinMaxScaler
if not df_cleaned.empty:
    df_cleaned = df_cleaned.copy()  # Ensure a fresh copy to avoid SettingWithCopyWarning
    df_cleaned[sensor_columns] = df_cleaned[sensor_columns].astype(np.float32)  # Convert first
    df_cleaned[sensor_columns] = scaler.fit_transform(df_cleaned[sensor_columns])  # Apply scaling

    print("✅ Sensor data normalized successfully.")
else:
    print("⚠️ Warning: No data available after outlier removal!")
```

```python
# Reset df_cleaned to original df
df_cleaned = df.copy()

# Confirm the reset
print("Data restored. Current shape:", df_cleaned.shape)
```

```python
from sklearn.preprocessing import StandardScaler
import numpy as np

# Step 2: Normalize the numeric columns before detecting outliers
scaler = StandardScaler()

# Selecting only numerical columns
numeric_columns = df_cleaned.select_dtypes(include=[np.number]).columns

# Apply StandardScaler
df_scaled = pd.DataFrame(scaler.fit_transform(df_cleaned[numeric_columns]), columns=numeric_columns)

# Compute Z-Scores on scaled data with a safer threshold
threshold = 5  # Keeping a higher threshold to avoid losing too much data
z_scores = np.abs(df_scaled)

# Apply the new threshold
df_cleaned = df_cleaned[(z_scores < threshold).all(axis=1)]

print("Outliers removed. New dataset shape:", df_cleaned.shape)
```

```python
from sklearn.preprocessing import MinMaxScaler

# Step 3: Apply Min-Max Scaling to normalize sensor data
scaler = MinMaxScaler()

# Selecting only numerical sensor columns (excluding categorical or index-like columns)
sensor_columns = [col for col in df_cleaned.columns if "sensor" in col]

# Apply Min-Max Scaling
df_cleaned[sensor_columns] = scaler.fit_transform(df_cleaned[sensor_columns])

print("✅ Sensor data normalized successfully.")
df_cleaned.head()
```

```python
# Step 4: Save the preprocessed data to a CSV file
df_cleaned.to_csv("preprocessed_data.csv", index=False)

print("✅ Preprocessed data saved successfully as 'preprocessed_data.csv'.")
```
